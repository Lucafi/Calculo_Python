## A equao geral da reta  definida por:

$$\frac{y-y_0}{x-x_0}=m$$

A reta tangente ao grfico da funo para $f(x)$ no ponto $(x_0, f(x_0))$

$$ \frac{y-f(x_0)}{x-x_0}=f'(x_0)$$

Desse modo, considerando a reta tangente a um ponto:

 
 
 $$ y= f'(x_0)(x-x_0)+f(x_0)$$

Desenvolvendo para  $x_0=a$


 $$ y= f'(a)(x-a)+f(a)$$

```python
a=2
f_a=-3*a**2 +7*a
f_a
```

```python
df_a=7-10*a
df_a
```

## Reta tangente que passa no ponto $x_0=4$



$$rt=f'(4)(x-4)+f(4)$$



$$rt=\frac{1}{4}x-1+2$$



$$rt=2+\frac{x-4}{4}$$

$f(x)=\sqrt(x)$, numa vizinhana de $x_0=4$

```python

plt.figure(2)
x=np.linspace(0,10,100)
y=np.sqrt(x)
rt=2+(x-4)/4
plt.plot(x,y)
plt.plot(x,rt)
```

```python
plt.plot(x,y)
plt.plot(x,rt)
plt.fill_between(x,y,rt, color='lime', alpha=0.3)
#plt.xlim(1.5,7)
```

## Polinmio de Taylor de ordem 2 na funo $f(x)$ no ponto $(x_0, f(x_0))$

$$P2(x)=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2}(x-x_0)^2$$

Aproximao de ordem 2 de $y=f(x)$


```python
x,y,d_r = symbols('x y \partial')
y=sqrt(x)
yd=diff(y)
d_r=diff(yd)
Derivative(d_r)                   
                   
                   
```

```python
d_r=d_r.subs(x,4)
```

```python
d_r
```

## Polinmio de Taylor de ordem 3 na funo $f(x)$ no ponto $(x_0, f(x_0))$

$$P2(x)=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2}(x-x_0)^2+\frac{f'''(x_0)}{3!}(x-x_0)^3$$

Aproximao de ordem 2 de $y=f(x)$



```python
x,y,d_r, d_2r = symbols('x y \partial \partial_2')
y=sqrt(x)
yd=diff(y)
d_r=diff(yd)
d_2r=diff(d_r)
Derivative(d_2r)                   

```

```python
d_2r=d_2r.subs(x,4)
```
