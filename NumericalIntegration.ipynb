# Integração Numérica

### A figura acima leva-nos a pensar que é possível resolver a área sob a curva descrita pela função $y=f(x)$, fazendo $A=\int_{a}^{b} y \ dx$. Também, conduz-nos a ilação de que se dividirmos a curva em numerosos trapézios teremos a área aproximada sob a curva, tal como:

### $$A=\int_{a}^{b} y \ dx\approx \frac{1}{2}(y_1+y_2)d+\frac{1}{2}(y_2+y_3)d+\frac{1}{2}(y_3+y_4)d+(y_n+y_{n+1})d$$

### Simplificando a expressão, temos:

### $$A=d\left[\frac{1}{2}y_1+y_2+y_3+y_4+...y_n+\frac{1}{2}y_{n+1}\right]$$



### Analíticamente podemos escrever que a área aproximada da função é:


$$A=\int_{a}^{b} y dx\approx \left(\frac{largura}{intervalo}\right)\left\{\frac{1}{2}\left(\frac{1^{a}+última}{ordenada}\right)+\left(\frac{soma de ordenadas}{remanescentes}\right)\right\}$$
    

### Determinação do intervalo

$$d=\frac{b-a}{n}$$

### onde $d$ é o intervalo
## $a$ e $b$ os limites da integração
## $n$ o número de ordenadas

### Seja a função $y=x^2+3x$

### Limites $a=-5$ e $b=5$

```python
x=np.linspace(-5,5,100000)
y=x**2+3*x
```

 ## LaTex

$$A=\int_{a}^{b} y dx\approx \left(\frac{largura}{intervalo}\right)\left\{\frac{1}{2}\left(\frac{1^{a}+última}{ordenada}\right)+\left(\frac{soma de ordenadas}{remanescentes}\right)\right\}$$

```python
a=-5
b=5
n=len(x)
d=(b-a)/n
```

```python
A=d*((y[0]+y[n-1])*0.5+np.sum(y)-y[0]-y[n-1])
```

```python
A
```

```python
print(f"A área sob a curva é {A:0.3f} unidades de área")
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y, color='lime', label='Função primária $x^2+3x$')
yi=x**3/3+3*x**2/2
plt.plot(x,yi, color='red', label='Função integrada $x^3/3+3x^2/2$')
plt.legend()
```

```python
yin=np.cumsum(y*d)
plt.plot(x,yi, color='red', label='Função integrada analítica $x^3/3+3x^2/2$')
plt.scatter(x,yin, color='blue', label='Função integrada numérica $x^3/3+3x^2/2$')
plt.legend()
```
