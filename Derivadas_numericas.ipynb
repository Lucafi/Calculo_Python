# Derivadas Numéricas - Diferenças Finitas

### A derivação numérica é feita a partir do conhecimento das séries de Taylor.  Há sempre três fórmulas:
## Primeira derivada

### a) Derivada à frente:

## $$\frac{d}{dx}f(x_{[i]}) = \frac{f(x_{([i+1]})-f(x_{[i]})}{\Delta x}$$

### b) Derivada atrás:

## $$\frac{d}{dx}f(x_{[i]}) = \frac{f(x_{[i]})-f(x_{[i-1]})}{\Delta x}$$

### c) Derivada central:

## $$\frac{d}{dx}f(x_{[i]}) = \frac{f(x_{[i+1]})-f(x_{[i-1]})}{2 \Delta x}$$



## Segunda derivada

### a) Derivada à frente:

## $$\frac{d^2}{dx^2}f(x_{[i]}) = \frac{f(x_{[i+2]})-2f(x_{[i+1]})+f(x_{[i]})}{\Delta x^2}$$

### b) Derivada atrás:

## $$\frac{d^2}{dx^2}f(x_{[i]}) = \frac{f(x_{[i]})-2f(x_{[i-1]})+f(x_{[i-2]})}{\Delta x^2}$$


### c) Derivada central:

## $$\frac{d^2}{dx^2 }f(x_{[i]}) = \frac{f(x_{[i+1]})-2f(x_{[i]})+f(x_{[i-1]})}{\Delta x^2}$$






# Exemplo

## $$f(x)=3x^3+2x^2+5x+1$$
## Limites $a=-2$ e $b=2$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=-2
b=2
x=np.linspace(a,b,100)
def f(x):
    return 3*x**3+2*x**2+5*x+1
y=f(x)

def df(x):
    return 9*x**2+4*x+5
y2=df(x)

def d2f(x):
    return 18*x+4
y3=d2f(x)
```

```python
plt.plot(x,y, color="coral", label="Função primária $y=3x^3+2x^2+5x+1$")
plt.plot(x,y2, color="olive", label="Função primeira derivada $y'=9x^2+4x+5$")
plt.plot(x,y3, color="steelblue", label="Função segunda derivada $y''=18x+4$")
plt.legend()
```

## Derivada Numérica

```python
N=1100
xi=np.linspace(a,b,N)
yi=3*xi**3+2*xi**2+5*xi+1
dx=(b-a)/(N-1)

yp=np.zeros_like(xi)

for i in range(N):
    if i==0:
        yp[i]=(yi[i+1]-yi[i])/dx
    elif i==N-1:
        yp[i]=(yi[i]-yi[i-1])/dx
    else:
        yp[i]=(yi[i+1]-yi[i-1])/(2*dx)
        
```

```python

ypp=np.zeros_like(xi)

for i in range (N):
    if i==0:
        ypp[i]=(yi[i+2]-2*yi[i+1]+ yi[i])/dx**2
    elif i==N-1:
        ypp[i]=(yi[i]-2*yi[i-1]+yi[i-2])/dx**2
    else:
        ypp[i]=(yi[i+1]-2*yi[i]+yi[i-1])/dx**2
```

```python
plt.plot(x,y2, color="olive", label="Função primeira derivada analítica $y'=9x^2+4x+5$")
plt.scatter(xi,yp, color="steelblue", label="Função primeira derivada numérica $yp=df[i]$")
plt.legend()
```

```python
plt.plot(x,y3, color="steelblue", label="Função segunda derivada $y''=18x+4$")
plt.scatter(xi,ypp, color="red", label="Função segunda derivada $ypp=df2[1}$")
plt.legend()
```
