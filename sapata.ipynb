# a) Ruptura Geral
$\phi=0$

```python
phi=0.01;
phi=(0.01*np.pi)/180
Nq=np.exp(np.pi*np.tan(phi))*np.tan(np.pi/4+phi/2)**2
```

```python
Nc=(Nq-1)*1/(np.tan(phi))
```

```python
print("O valor de Nc √© %.2f"% Nc)
```

$$\sigma_r=c.N_c.S_c+q.N_q.Sq$$

```python
B=2
L=3
Sc=1+(B/L)*(Nq/Nc)
print("O valor de Sc √© %.2f"%Sc)
```

```python
Sq=1+np.tan(phi)
print("O valor de Sq √© %.2f"%Sq)
```

```python
N=15
c=10*N
print("O valor de c √© %.2f"%c)
```

```python
gamma=19
h=1
q=gamma*h
print("O valor de q √© %.2f"%q)
```

```python
Tr1=c*Nc*Sc+q*Nq*Sq
Tr1=Tr1/1000
print("A tens√£o de ruptura geral √© %.2f"%Tr1,"MPa")
```
